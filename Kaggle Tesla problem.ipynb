```python
#Train and test dataset
from sklearn.model_selection import train_test_split
```

```python
#set x and y data
x=Tesla_dataset[['High','Low','Close']]
y=Tesla_dataset['Volume']
```

```python
#check head of x
x.head()
```

```python
#check head of y
y.head()
```

```python
x.info()
```

```python
#trane First 30% data send to testting
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.30)
```

```python
X_test.head()
```

```python
#Import LinearRegression
from sklearn.linear_model import LinearRegression
```

```python
#create object linearregeression
Teslaregression = LinearRegression()
```

```python
#send test data to LinearRegression object
Teslaregression.fit(X_train, y_train)
```

```python
#predict
Teslaregression.predict([[75, 250,10]])
```

```python
#Stor all Predict datato ypredict object
y_predict = Teslaregression.predict(X_test)
```

```python
#check is stor or not
y_predict[2]
```

```python
# Genrate Scatter Graph
plt.title=("Tesla Prediction")
plt.xlabel("X label")
plt.ylabel("Y label")

plt.scatter(y_predict,y_test, c=['r'], alpha = 0.2, marker=r'$\clubsuit$')
```

```python
# More understand to use color in Scatter graph
N=3
plt.title=("Tesla Prediction")
plt.xlabel("X label")
plt.ylabel("Y label")
colors = np.random.rand(N)
plt.scatter(y_predict,y_test, c=colors, alpha = 0.8, marker=r'$\clubsuit$')
plt.show()
```
